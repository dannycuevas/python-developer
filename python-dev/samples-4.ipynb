# Functions
- Python "**Functions** allow us to create blocks of code that can be easily executed many times, without needing to constantly rewrite the entire block of code"

```python
def sum_numbers(num1,num2):
  return num1+num2

result = sum_numbers(2,3)
print(result)
```

```python
# set up a "default" value at the Variable level, in case you forget to pass in an argument
# in this example, our "default" value will be "Batman"
# notice how when executing the Function, there is no Value being pass-in in the parenthesis
def say_hello(thename="Batman"):
  print(f"Hello there, {thename}")

say_hello()
```
