```python
# This a List that multiplies each value of each List with the next List
# And then appends the results of each operation of X and Y, to the original List
mylist = []

for x in [2,4,6]:
  for y in [1,10,1000]:
    mylist.append(x*y)

print(mylist)
```

```python
# From a String, print out the words that start with letter "s"
# Grab the first letter of each word with index 0 that is equal to "s"
# Make sure that is case-insesitive 
palabra = 'Sam name - Print only the words that start with s in this sentence'

for word in palabra.split():
  if word[0] == 's' or word[0] == 'S':
    print(word)
```
