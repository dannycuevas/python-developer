- This next example is a basic Function that will check if a number is an even number
- It will accomplish the task by returning the "mod operator" equal to zero (% 2)

```python
# we are building a Function that can check if a number is even
# this is done by using "mod 2" or % 2, for a remainder of zero 0
def even_number(number): # define the Function and give it a name, and define the parameter
  return number % 2 == 0 # give me back (return) that parameter (argument - number) with mod-2, and having to equal zero (==)
```

```python
# if the reminder of the division is zero, then it will be True
even_number(22)
```

```python
# return True if any number is even inside of a List
# the parameter we are passing-in is assumed to be the List
def even_list(num_list):
  for number in num_list:   # check FOR every item/number inside my object (this List):
    if number % 2 == 0:   # if any of those items/numbers mod-2 is equal to zero:
      return True   # then return True:
    else:   # if none found, then:
      pass   # keyword "pass", so do not do anything
```

```python
# first we will pass-in a List with odd numbers only, to test the Function
even_list([1,3,5])
```

```python
# now we pass-in a List with even numbers to make sure it works
even_list([2,4,5])
```

- The next version of the Function will actually print out "False" when there are no even numbers
- To do this, we do not use the final "return" at the IF statement identation, since that would mean that there is a "return" call on either condition of the IF statement, because, essentially, that would mean it is only checking the first item/number condition
- Solution: we want to check every single item/number for the main condition and make sure whether or not it is an even number before "returning false"
- The "return False" should actually be an indentation with the For loop

```python
def even_list(num_list):
  for number in num_list:   # check FOR every item inside my object (this List)
    if number % 2 == 0:   # if any item mod 2, is equal to zero
      return True   # then return True
    else:   # if none found, then
      pass   # keyword "pass", so do not do anything
# after checking either condition in all the items/numbers with the For loop, then this says:
# "if the For loop has completed, and we never broke out of the Function with the first return statement, then go ahead and execute the following:"
  return False   
```

- For the next project, now we want to "return all even numbers in a List"
- And if there are no even numbers, then just return an empty List

```python
def even_list(num_list):
  even_numbers = []   # empty placeholder List, since we will want to return something later

  for number in num_list: # check FOR every item/number inside my object (this List):
    if number % 2 == 0: # if any item/number mod 2, is equal to zero:
      even_numbers.append(number) # then append that item/number to my empty List from above
    else:
      pass
  # once the Function has iterated the entire List with the For loop, we can return that new List
  return even_numbers 
```

```python
# we now run the Function to print a List of even numbers found in another List
even_list([1,2,3,4,5,6,7,8])
```

- Returning multiple items from a Function with Tuple unpacking

```python
# creating an example with "stock prices"
stock_prices = [('apple', 200), ('goog',400),('msft', 100)]
```

```python
# for every item in my Tuple, print every item
# this will print each parenthesis as a single object
for item in stock_prices:
  print(item)
```

```python
# now we do "Tuple unpacking", meaning to individually grab the items from the Tuples
# to do this, we will use 2 Variables, so each variable can call each item that is inside the parenthesis, for example, just retrieving the first item of each Tuple as "ticket"
for ticket,price in stock_prices:
  print(ticket)
```

```python
# now we get only the"prices" in each Tuple
# but for each price, we will get them them with a 10% increase in price, this works by getting the price, plus, the price by 0.1 + price
# we still have to call 2 spaces of Variables (like 2 items inside the Tuples)
for ticket,price in stock_prices:
  print(price+(0.1*price))
```

- So far, that is how a For loop works to return the Tuples values, but we can do the same thing with Functions
- A Function can actually return things as Tuples, and then we can unpack the result from the Function

```python
# these tuples will have employees names, and hours they have work for the 
# then we want to see who the employee of the month is, by finding out the employee with the most hours worked
work_hours = [('Abby', 100), ('Billy', 400), ('Dan', 2000), ('Cass', 800)]
```

```python
# our Function should check all of the names and hours, and keep track of the name with the most hours
# this Function will accept the list of Tuples as the argument then
# we want to have a placeholder Variables at the top of the Function
def employee_check(work_hours):
  current_max = 0
  employee_of_month = ""

# so then the For loop will have to do some Tuple unpacking
  for employee,hours in work_hours: # for every emp-hours, check:
    if hours > current_max: # if emp hours are greater than my current_max of hours, then:
      current_max = hours # reset my current_max to the hours of that particular emp, and then:
      employee_of_month = employee # employee_of_month is equal to the emp name in that Tuple
    else: # otherwise,
      pass # simply pass  

# at the end, want to return a Tuple with the Name and hours
  return (employee_of_month, current_max)
```

```python
employee_check(work_hours)
```
