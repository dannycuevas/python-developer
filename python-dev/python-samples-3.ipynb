# Examples of FOR loops

```python
mylist = [1,2,3,4,5] #this is a "statement" (or assigment)
for jelly in mylist:
  print(jelly)
```

```python
for number in mylist:
  if number % 2 == 0: # Check for "even numbers" only
    print(number)
  else:
    print(f"Odd numner: {number}")  # Check for "odd numbers" only
```
