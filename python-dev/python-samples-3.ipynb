# FOR Loops
- **Loops are "constructs" that enable programs (or block of codes) to execute instructions more than once, so you do not have to copy paste that code N number of times, and also can be used to iterate over sequences of items in Python**
	- "Repeat a section of code" is also called "iteration"... Or just informally, just a "loop"
	- Many objects in Python are "iterable" - meaning, we can iterate over every element in the "object" (such as every element in a List, or every character in a String)
	- We can use FOR loops to execute a block of code for every iteration (example, to print out every single letter in a String)

```python
mylist = [1,2,3,4,5] #this is a "statement" (or assigment)
for jelly in mylist:
  print("hello Sabuza")
```

#### Modulo
The "modulo" allows us to get the remainder in a division, and uses the % symbol, example:

```python
# 15 divided by 5 is 3, with a remainder of 2:
17 % 5
```

```python
# Example of a FOR loop, by printing both even and odd numbers
for number in mylist:
  if number % 2 == 0: # checks for even numbers
    print(number)
  else:
    print(f"Odd number: {number}")  # f-statement that prints out the odd numbers
```

```python
list_sum = 0 # starts off at zero

for number in mylist:
  list_sum = list_sum + number
  print(list_sum)

# then it goes "zero + 1 = 1" and gets printed out
# then goes "1 + 2 = 3" and gets printed out
# then "3 + 3 = 6"
# then "6 + 4 = 10"
# then 10 + 5 = 15
```

# WHILE Loops
- A While Loop is different from a For Loop, in that we way "while a condition is happening; then do something"

```python
x = 0

while x < 5: # while X is less than 5, do some action
  print(f"The current value of x is {x}") # will print infinetely since X is not changing
  x = x + 1 # X + 1 every time the loop runs - while the condition is True
```

```python
x = 0

while x < 5: # while X is less than 5, do some action
  print(f"The current value of x is {x}") # will print infinetely since X is not changing
  x = x + 1 # X + 1 every time the loop runs - while the condition is True
else: # adding an Else statement
  print("X is no longer less than 5")
```

```python
# Example of "continue" statement
mystring = "Sammy"
for letter in mystring:
  if letter == 'a':
    continue
  print(letter)
```

```python
# Example of "break" statement
x = 0
while x < 5:
  if x == 4:
    break
  print(x)
  x = x + 1
```

```python
# Using "list comprehension"
mylist_two = [x for x in "batman"]
print(mylist_two)
```

```python
# Convert this range into a list, but only with even numbers
mylist = [x for x in range(0,11) if x%2==0]
print(mylist)
```

```python
# Convert this range into a list, but only with even numbers
# And we then, want the square of the even numbers
mylist = [x**2 for x in range(0,11) if x%2==0]
print(mylist)
```

```python
# Converting Celsius to F
celsius = [0,10,20,34,34.5]
# Formula for the conversion
fahrenheit = [((9/5)*temp + 32) for temp in celsius]
print(fahrenheit)
```

```python
# Converting Celsius to F - using a For loop
celsius = [0,10,20,34,34.5]
fahrenheit = []

for temp in celsius:
	fahrenheit.append(((9/5)*temp + 32))
  
print(fahrenheit)
```

**Write a program that prints the integers from 1 to 100. But for multiples of three print "Fizz" instead of the number, and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".**

```python
# print out the integers from 1 up to 101, excep we are replacing integers with words
# we print out these words when there are multiples of 3, multiples of 5, or both of 3 and 5
# we are printing out FizzBuzz as the first IF statement, as it has 2 conditions to be met using "And"
# when no words, print the integer itself (under the Else statment)
for num in range(1,101):
  if num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")
  elif num % 3 == 0:
    print("Fizz")
  elif num % 5 == 0:
    print("Buzz")
  else:
    print(num)      
```
