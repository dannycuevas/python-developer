# Project: ChatBot

- This will be a very simple chatbot 
- We will just add one extra feature to allow the bot to sum numbers for us
- Recommending to not run this in a Jupyter notebook, maybe Spyder IDE a better option

```python
bot_name: str = "Izuna"
print(f"Hello-des!... this is Izuna-des... how can I help you-des?")

# now we start our While loop, which will be an infinite loop
# and the first thing we will do, is take some user input

while True:
  user_input: str = input("You: ").lower() # lowercase all user input

  if user_input in ["hi", "hello"]:
    print(f"{bot_name}: hi-des!, how can I help you-des?")
  elif user_input in ["bye", "later", "see you"]:
    print(f"{bot_name}: mataneee-des!")
  # now we can make it so out bot can perform mathematical functions
  elif user_input in ["+", "add"]:
    print(f"{bot_name}: Hai-des!, please enter 2 numbers wapo-kun:")
    try:
      num1: float = float(input("First number-des: "))
      num2: float = float(input("Second number-des: "))
      print(f"{bot_name}: Hai-des!, the sum total is {num1 + num2}")
      # and if it does not work, we will add the "Except" block
      # this is in the case the user enters something that is not a number
    except ValueError:
      print(f"{bot_name}: Gomenasai!, that is not a valid number-des!, try again-des!")
      # now we go back to the if condition statement
  else:
    # and this is going to cover for all the cases we did not cover for above
    print(f"{bot_name}: Gomenasaiiii!, I do not understand-des y.y Please try again wapo-kun!")
```
